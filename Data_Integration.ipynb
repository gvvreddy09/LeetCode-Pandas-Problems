## Replace Employee ID With The Unique Identifier

```python
employees_dict = {
    'id': [1,7,11,90,3],
    'name': ['Alice', 'Bob', 'Meir', 'Winston', 'Jonathan']
}

employeeuni_dict = {
    'id': [3,11,90],
    'unique_id': [1,2,3]
}

employees = pd.DataFrame(employees_dict)
employeeuni = pd.DataFrame(employeeuni_dict)

print(employees)
print(employeeuni)
```

Write a solution to show the unique ID of each user, If a user does not have a unique ID replace just show null.

```python
pd.merge(employees, employeeuni, how='left', on='id')[['unique_id', 'name']]
```

## Students and Examinations

```python
students_dict = {
    'student_id':[1,2,13,6],
    'student_name':[None, 'Bob', 'John', 'Alex']
}

subjects_dict = {'subject_name':['Maths', 'Physics', 'Programming']}

examinations_dict = {
    'student_id':[1,1,1,2,1,1,13,13,13,2,1],
    'subject_name':['Maths', 'Physics', 'Programming', 'Programming', 'Physics', 'Maths', 'Maths', 'Programming', 'Physics', 'Maths', 'Maths']
}  

students = pd.DataFrame(students_dict)
subjects = pd.DataFrame(subjects_dict)
examinations = pd.DataFrame(examinations_dict)

print(students)
print(subjects)
print(examinations)
```

Write a solution to find the number of times each student attended each exam.

```python
examinations['flag'] = 1
```

```python
# Students-Subjects
cross_merging = pd.merge(students, subjects, how='cross')
cross_merging
```

```python
merged = pd.merge(cross_merging, examinations, how='left', on=['student_id', 'subject_name'])
merged
```

```python
merged.groupby(['student_id', 'student_name', 'subject_name'], as_index=False, dropna=False).agg(
    attended_exams = ('flag', 'sum')
)
```

## Actors and Directors Who Cooperated At Least Three Times

```python
actor_director_dict = {
    'actor_id': [1,1,1,1,1,2,2],
    'director_id': [1,1,1,2,2,1,1],
    'timestamp': [0,1,2,3,4,5,6]
}

actor_director = pd.DataFrame(actor_director_dict)

actor_director
```

Write a solution to find all the pairs (actor_id, director_id) where the actor has cooperated with the director at least three times.

```python
# My Approach
grouped = actor_director.groupby(['actor_id', 'director_id'], as_index=False).agg(
    cnt = ('timestamp', 'count')
)
grouped[['actor_id', 'director_id']][grouped['cnt']>=3]
```

The query method is used for efficient filtering of rows where the count (cnt) is greater than or equal to 3.

```python
# Improved version
actor_director.groupby(['actor_id', 'director_id'], as_index=False).agg(cnt = ('timestamp', 'count')).query('cnt >= 3')[['actor_id', 'director_id']]
```

## Sales Person

```python
SalesPerson_dict = {
    'sales_id': [1,2,3,4,5],
    'name': ['john', 'amy', 'mark', 'palm', 'alex'],
    'salary': [100000, 12000, 65000, 25000, 5000],
    'commission_rate':[6, 5, 12, 25, 10],
    'hire_date': [datetime(2006, 4, 1), datetime(2010, 5, 1), datetime(2008, 12, 25), datetime(2005, 1, 1), datetime(2007, 2, 3)]
}

Company_dict = {
    'com_id': [1,2,3,4],
    'name': ['RED', 'ORANGE', 'YELLOW', 'GREEN'],
    'city': ['Boston', 'New York', 'Boston', 'Austin']
}

Orders_dict = {
    'order_id': [1,2,3,4],
    'order_date': [datetime(2014, 1, 1), datetime(2014, 1, 2), datetime(2014, 1, 3), datetime(2014, 1, 4)],
    'com_id': [3,4,1,1],
    'sales_id': [4,5,1,4],
    'amount': [10000, 5000, 50000, 25000]
}

sales_person = pd.DataFrame(SalesPerson_dict)
company = pd.DataFrame(Company_dict)
orders = pd.DataFrame(Orders_dict)

print(sales_person)
print(company)
print(orders)
```

Write a solution to find the names of all the salespersons who did not have any orders related to the company with the name "RED".

```python
# Merge Orders & Company
OrdersCompany = pd.merge(orders, company, how='inner', on='com_id').rename(columns={'name':'com_name'})

# Merge OrdersCompany & SalesPerson
merged = pd.merge(OrdersCompany, sales_person, how='right', on='sales_id')

# Approach 2

# Get Sales Persons who are associated with the company RED and convert it to set type (handles duplicates)
sales_person_in_red_company = set(merged[merged['com_name'] == 'RED']['name'])

# Get all the Sales Persons from merged data frame and convert it to set type
sales_persons = set(merged['name'])

print(sales_person_in_red_company)
print(sales_persons)

# Get the Sales Persons who are not associated with the company RED by performing difference between the above two sets
pd.DataFrame(sales_persons - sales_person_in_red_company, columns=['name'])
```

```python
# Approach 1

# Get Sales Persons who are associated with the company RED and convert it to List 
sales_person_in_red_company = list(merged[merged['com_name'] == 'RED']['name'].unique())
sales_person_in_red_company
```

<span style="font-family:Comic Sans MS; color:red">[**isin**](https://www.w3schools.com/python/pandas/ref_df_isin.asp) documentation</span>

The isin() method checks if the Dataframe contains the specified value(s).

It returns a DataFrame similar to the original DataFrame, but the original values have been replaced with True if the value was one of the specified values, otherwise False.

```python
# Using .isin method Get the Sales Persons who are not associated with the company RED
pd.DataFrame(merged[~merged.isin(sales_person_in_red_company)]['name'].unique(), columns=['name']).dropna()
```

## Managers with at Least 5 Direct Reports

```python
employee_dict = {
    'id': [101,102,103,104,105,106],
    'name': ['john', 'dan', 'james', 'ammy', 'ane', 'ron'],
    'department': ['a', 'a', 'a', 'a', 'a', 'b'],
    'manager_id': [None, 101, 101, 101, 101, 101]
}

employee = pd.DataFrame(employee_dict)
employee
```

Write a solution to find managers with at least five direct reports.

```python
result = pd.merge(employee, employee, how='inner', left_on='id', right_on='manager_id', suffixes=('_manager','_employee'))[
['id_manager', 'name_manager', 'id_employee', 'name_employee']].groupby('id_manager', as_index=False).agg({'id_employee':'count'}).rename(columns={
    'id_manager':'manager_id', 'id_employee':'reportees'
})

result
```

```python
merge_back = pd.merge(result, employee, how='inner', left_on='manager_id', right_on='id')[['id','name','reportees']]

merge_back[['name']][merge_back['reportees']>=5]
```
